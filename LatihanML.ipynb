```python
### Looping 

words = ['cat', 'window', 'defenestrate']

for w in words:
    print(w, len(w))


users = {'Hans': 'active', 'Éléonore': 'inactive', '景太郎': 'active'}

for user, status in users.copy().items():
    if status == 'inactive':
        del users[user]

active_users = {}

for user, status in users.items():
    if status == 'active':
        active_users[user] = status

print(active_users)
```

```python
## Modul 4: Visualisasi Data
### Visualisasi data dilakukan menggunakan metode scatter plot sesuai modul, kemudian ditambahkan visualisasi lain untuk melihat distribusi data.

import pandas as pd
import matplotlib.pyplot as plt

# membaca dataset
data = pd.read_csv("tips.csv")

# Scatter Plot (sesuai modul)
plt.title("Scatter Plot Tip vs Day")
plt.xlabel("Day")
plt.ylabel("Tip")

plt.scatter(data['day'], data['tip'])

plt.show()


# Visualisasi tambahan: Bar Chart jumlah tips per hari
tips_per_day = data.groupby('day')['tip'].sum()

plt.title("Total Tip Berdasarkan Hari")
plt.xlabel("Day")
plt.ylabel("Total Tip")

plt.bar(tips_per_day.index, tips_per_day.values)

plt.show()
```

```python
## Eksplorasi Dataset Baru: Sales and Customer Insights

###Dataset ini berisi informasi mengenai pelanggan dan pengeluaran pelanggan. 
###Analisis dilakukan untuk memahami pola perilaku pelanggan berdasarkan usia, gender, dan total pengeluaran.

import pandas as pd
import matplotlib.pyplot as plt

# Membaca dataset
sales = pd.read_csv("sales_and_customer_insights.csv")

# Menghapus kolom yang tidak diperlukan
sales = sales[['Customer_ID',
               'Product_ID',
               'Purchase_Frequency',
               'Average_Order_Value',
               'Most_Frequent_Category',
               'Region']]

# Menampilkan beberapa data
print("5 Data Pertama:")
print(sales.head())


# -----------------------------------
# Visualisasi 1: Distribusi Frekuensi Pembelian
# -----------------------------------
plt.title("Distribusi Frekuensi Pembelian Pelanggan")

plt.hist(sales['Purchase_Frequency'])

plt.xlabel("Purchase Frequency")
plt.ylabel("Frekuensi")

plt.show()


# -----------------------------------
# Visualisasi 2: Distribusi Kategori Produk
# -----------------------------------
category_count = sales['Most_Frequent_Category'].value_counts()

plt.title("Distribusi Kategori Produk")

plt.pie(category_count,
        labels=category_count.index,
        autopct='%1.1f%%',
        startangle=90)

plt.show()


# -----------------------------------
# Visualisasi 3: Hubungan Frekuensi Pembelian dan Nilai Pesanan
# -----------------------------------
plt.title("Hubungan Frekuensi Pembelian dan Nilai Pesanan")

plt.scatter(sales['Purchase_Frequency'], sales['Average_Order_Value'])

plt.xlabel("Purchase Frequency")
plt.ylabel("Average Order Value")

plt.show()




```
